```python
from bs4 import BeautifulSoup
import requests
import os
```

```python
url='https://www.searchenginejournal.com/elon-musks-twitter-takeover-a-timeline-of-events/470927/'

html = requests.get(url)
```

```python
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
soup
```

```python
result = soup.find('section', class_='sec-art')

```

```python
queries = result.find_all('h2')

queries
```

```python
queries2 = result.find_all('p')

queries2
```

```python
for query in queries:

    try:
        queries = result.find_all('h2')
        queries2 = result.find_all('p')
        
        if(queries and queries2):
            print('\n-----------------\n')
            print(queries)
            print(queries2)
    except AttributeError as e:
        print(e)
```
